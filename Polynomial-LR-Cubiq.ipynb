```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Data
#Scikit-Learn Requirement: Many machine learning libraries, including Scikit-Learn, 
#expect the input data to be in a 2D format (i.e., each feature is in its own column, and each sample is in a separate row). 
#reshape(-1, 1) turns the 1D array [1, 2, 3, 4, 5, 6, 7] into a 2D array

years_of_experience = np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1, 1)
salary = np.array([50000, 55000, 65000, 80000, 110000, 150000, 200000])

# Polynomial regression with degree 2
poly = PolynomialFeatures(degree=2)
years_poly = poly.fit_transform(years_of_experience)

# Linear regression model
model = LinearRegression()
model.fit(years_poly, salary)

# Predictions
salary_pred = model.predict(years_poly)

# Plotting
plt.scatter(years_of_experience, salary, color='blue', label='Actual data')
plt.plot(years_of_experience, salary_pred, color='red', label='Polynomial regression (degree=2)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()

# Print coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Data
years_of_experience = np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1, 1)
salary = np.array([50000, 55000, 65000, 80000, 110000, 150000, 200000])

# Polynomial regression with degree 3
poly = PolynomialFeatures(degree=3)
years_poly = poly.fit_transform(years_of_experience)

# Linear regression model on transformed polynomial features
model = LinearRegression()
model.fit(years_poly, salary)

# Predictions
salary_pred = model.predict(years_poly)

# Plotting
plt.scatter(years_of_experience, salary, color='blue', label='Actual data')
plt.plot(years_of_experience, salary_pred, color='red', label='Cubic regression (degree=3)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()

# Print the model's coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Data
years_of_experience = np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1, 1)
salary = np.array([50000, 55000, 65000, 80000, 110000, 150000, 200000])

# Define function to fit and evaluate models
def fit_and_evaluate(degree):
    poly = PolynomialFeatures(degree=degree)
    years_poly = poly.fit_transform(years_of_experience)
    model = LinearRegression()
    model.fit(years_poly, salary)
    salary_pred = model.predict(years_poly)
    mse = mean_squared_error(salary, salary_pred)
    r2 = r2_score(salary, salary_pred)
    
    print(f"Degree {degree} Polynomial Regression:")
    print("Intercept:", model.intercept_)
    print("Coefficients:", model.coef_)
    print("Mean Squared Error:", mse)
    print("R-squared:", r2)
    print("-" * 50)

# Linear Regression (Degree 1)
fit_and_evaluate(degree=1)

# Polynomial Regression (Degree 2)
fit_and_evaluate(degree=2)

# Cubic Regression (Degree 3)
fit_and_evaluate(degree=3)
```

* Intercept and Coefficients: As we move from linear to quadratic to cubic, additional terms improve the models ability to capture the complexity of the salary vs. experience relationship.
* The linear model only captures a simple upward trend, while the quadratic and cubic models capture a curving relationship, with the cubic model adding even more nuance.
* Mean Squared Error: The MSE decreases as we increase the degree, indicating that the model fit improves with more flexibility in capturing the datas pattern.
* R-squared: R-squared increases with higher-degree models, with the cubic model explaining nearly all of the variance in the data.
